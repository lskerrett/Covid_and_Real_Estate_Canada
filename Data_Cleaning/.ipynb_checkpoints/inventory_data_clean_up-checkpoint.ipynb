```python
import pandas as pd
import datetime as dt
```

```python
housing_inventory=pd.read_csv("../Resources/Housing_inventory.csv", low_memory=False)
```

```python
housing_inventory.columns.tolist()
```

```python
housing_inventory.dtypes
```

```python
housing_inventory = housing_inventory.drop(["TERMINATED", "SYMBOL","DECIMALS", "DGUID","STATUS","UOM","UOM_ID","SCALAR_FACTOR","SCALAR_ID","VECTOR","COORDINATE"], axis=1)
```

```python
housing_inventory = housing_inventory.rename(columns = {"VALUE":"UNIT_VALUE","Type of dwelling unit":"Dwelling_Type","Completed dwelling units":"Completed_units"})
housing_inventory.head()
```

```python
housing_inventory["REF_DATE"] = housing_inventory["REF_DATE"].astype("datetime64")
housing_inventory["GEO"] = housing_inventory["GEO"].astype("str")
housing_inventory["UNIT_VALUE"] = housing_inventory["UNIT_VALUE"].astype("int")
```

```python
housing_inventory.dtypes
```

```python
housing_inventory=housing_inventory.dropna()
```

```python
housing_inventory["GEO"].nunique()
```

```python
housing_inventory=housing_inventory[housing_inventory.GEO!="Census metropolitan areas"]
housing_inventory=housing_inventory[housing_inventory.Dwelling_Type!="Total units"]
housing_inventory = housing_inventory.reset_index()
housing_inventory.head()
```

```python
housing_inventory.drop(columns = ["index"], inplace = True)
housing_inventory.head()
```

```python
df=housing_inventory.GEO.str.split(", ",expand=True,)
df.head()
```

```python
df = df.drop(columns = [2])
df
```

```python
index_list = df.dropna().index
index_list
```

```python
df = df.loc[index_list,:].reset_index()
df.head()
```

```python
df.drop(columns = ["index"], inplace=True)
df.head()
```

```python
housing_inventory = housing_inventory.loc[index_list,:].reset_index()
housing_inventory.head()
```

```python
housing_inventory.drop(columns = ["index"], inplace = True)
housing_inventory.head()
```

```python
housing_inventory_df = pd.concat([housing_inventory,df], axis=1)
housing_inventory_df.head()
```

```python
housing_inventory_df = housing_inventory_df.rename(columns = {0:"City",1:"Province"})
housing_inventory_df.head()
```

```python
housing_inventory_df=housing_inventory_df.drop(columns={'GEO'})
housing_inventory_df.head()
```

```python
cols=['REF_DATE','City','Province','Completed_units','Dwelling_Type','UNIT_VALUE']
housing_inventory_df=housing_inventory_df[cols]
housing_inventory_df.head()
```

```python
housing_inventory_df=housing_inventory_df.set_index("REF_DATE")
housing_inventory_df.head()
```

```python
housing_inventory_df.to_csv("../Resources/Housing_inventory_clean.csv",index=True)
```
